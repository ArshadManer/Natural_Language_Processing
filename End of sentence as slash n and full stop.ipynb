```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
text = 'Guest Blogging. Guest Blogging allows the user to collect backlinks.'
```

```python
doc=nlp(text)
```

```python
for ele in doc.sents:
    print(ele.text)
```

```python
config = {"punct_chars": [',']}
nlp.add_pipe("sentencizer",before="parser",config=config)
for sent in nlp(text).sents:
    print(sent)
```
