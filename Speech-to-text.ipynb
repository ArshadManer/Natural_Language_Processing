```python
import speech_recognition as sr
import spacy

nlp = spacy.load("en_core_web_sm")

def Speech_to_text(Audio_file_path):
    recognizer = sr.Recognizer()
    try:
        with sr.AudioFile(Audio_file_path) as source:
            print("Processing audio...")
            audio = recognizer.record(source)
            text = recognizer.recognize_google(audio)
            print("Speech to Text Result:")
            print(text)
            
            doc = nlp(text)
            print("\nLemmas:")
            for token in doc:
                print(f"{token.text} --> {token.lemma_}")

    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand audio")
    except sr.RequestError as e:
        print(f"Could not request results from Google Speech Recognition service; {e}")
        
file_path = "C:\Download\OSR_us_000_0030_8k.wav"
Speech_to_text(file_path)
```

```python
import speech_recognition as sr
import spacy
import pandas as pd
import zipfile
import os
import urllib.request

nlp = spacy.load("en_core_web_sm")

def Speech_to_text(Audio_file_path):
    recognizer = sr.Recognizer()
    try:
        with sr.AudioFile(Audio_file_path) as source:
            print("Processing audio...")
            audio = recognizer.record(source)
            text = recognizer.recognize_google(audio)
            print("Speech to Text Result:")
            print(text)
            
            doc = nlp(text)
            print("\nLemmas:")
            for token in doc:
                print(f"{token.text} --> {token.lemma_}")

    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand audio")
    except sr.RequestError as e:
        print(f"Could not request results from Google Speech Recognition service; {e}")

def download_and_extract_data(github_url, target_dir):
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)
    
    try:
        filename, headers = urllib.request.urlretrieve(github_url, os.path.join(target_dir, "Stock_data.zip"))
        with zipfile.ZipFile(os.path.join(target_dir, "Stock_data.zip"), 'r') as zip_ref:
            zip_ref.extractall(target_dir)
        print("Data downloaded and extracted successfully.")
    except Exception as e:
        print(f"An error occurred while downloading or extracting the data: {e}")

if __name__ == "__main__":
    github_url = input("Enter the GitHub link for the dataset: ")
    target_dir = "data"  # You can change this directory name if needed

    download_and_extract_data(github_url, target_dir)

    audio_file_path = input("Enter the path to the audio file: ")
    Speech_to_text(audio_file_path)
```
